```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
dry = cv2.imread('../Images/Dry03.jpeg')
dry = cv2.cvtColor(dry, cv2.COLOR_BGR2RGB)

wet = cv2.imread('../Images/Wet03.jpeg')
wet = cv2.cvtColor(wet,cv2.COLOR_BGR2RGB)
```

```python
cv2.imshow("Dry Image", cv2.cvtColor(dry, cv2.COLOR_BGR2RGB))
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
dry = np.array(dry, dtype = np.int64)
wet = np.array(wet, dtype = np.int64)
```

```python

def extract_color(x):
    """ 
        Use this function get separate there color channels thats mean there different data frame as red, green, blue 
        Input - numpy array of a color image
    """
    x_red = x.copy()
    x_red[:,:,1:3] = 0

    x_green = x.copy()
    x_green[:,:,0:3:2] = 0

    x_blue = x.copy()
    x_blue[:,:,0:2] = 0
    return x_red, x_green, x_blue


def get_row_data(red, green,blue,row_num):
    """
        This function will return a single dataframe of a particular Row number 
        Input - red_df, green_df, blue_df, row_num 
        output- A single datafram wit column name Length, Red, Green , Blue 
    """
    df = pd.DataFrame({'Length':np.arange(0,red.shape[1]), 'Red':red[row_num,:,0], 'Green': green[row_num,:,1], "Blue":blue[row_num,:,2]})
    return df

def just_two_row_plot(df1, df2, color = 'Blue'):
    plt.figure(figsize=(20,20))
    plt.plot(df1[color], c = 'red')
    plt.plot(df2[color], c = 'blue')
    plt.legend(['dry', 'wet'])
    fig_name = 'Row-Average-5' + color
    plt.savefig(f'{fig_name}.jpg', dpi = 1000)
    plt.show()
    
def two_row_with_sma(df1, df2, color='Blue', sma_vlaue = 250):
    column_name =  'SMA-'+str(sma_vlaue)
    df1[column_name] = df1[color].rolling(sma_vlaue, center=True).mean()
    df2[column_name] = df2[color].rolling(sma_vlaue, center = True).mean()
    
    plt.figure(figsize=(20,5))
    plt.plot(df1[column_name],c='r')
    plt.plot(df2[column_name],c='b')
    plt.legend(['df1-smooth','df2-smooth'])
    plt.savefig('df1 vs df2 with sma.jpg',dpi = 1000)
    plt.show()
```

```python
dry_red, dry_green, dry_blue = extract_color(dry)
wet_red, wet_gree, wet_blue = extract_color(wet)
```

```python
row_num = 150
dry_df = get_row_data(dry_red, dry_green, dry_blue, row_num=row_num)
wet_df = get_row_data(wet_red, wet_gree, wet_blue, row_num= row_num)
```

```python
just_two_row_plot(dry_df, wet_df, color='Blue')
```

```python
two_row_with_sma(dry_df, wet_df, color='Red', sma_vlaue=350)
```

## Taking Row Averages

```python
dry_red, dry_green, dry_blue = extract_color(dry)
wet_red, wet_gree, wet_blue = extract_color(wet)

# row_nums = np.linspace(50,688, 5, dtype=np.int64)
row_num = 250
dry_df = get_row_data(dry_red, dry_green, dry_blue, row_num=row_num)
wet_df = get_row_data(wet_red, wet_gree, wet_blue, row_num= row_num)
```

```python
row_nums = np.linspace(50,688, 5, dtype=np.int64)

dfs1 = []
dfs2 = []
for i in row_nums:
    dfs1.append(get_row_data(dry_red, dry_green, dry_blue, row_num=i))
    dfs2.append(get_row_data(wet_red, wet_gree, wet_blue, row_num= i))
```

```python
r1 = 0
g1 = 0
b1 = 0

r2 = 0
g2 = 0
b2 = 0

for i in range(len(dfs1)):
    r1 = r1 + dfs1[i]['Red']
    g1 = g1 + dfs1[i]['Green']
    b1 = b1 + dfs1[i]['Blue']

    r2 = r2 + dfs2[i]['Red']
    g2 = g2 + dfs2[i]['Green']
    b2 = b2 + dfs2[i]['Blue']
    
r1 = r1/len(dfs1)
g1 = g1/len(dfs1)
b1 = b1/len(dfs1)

r2 = r2/len(dfs1)
g2 = g2/len(dfs1)
b2 = b2/len(dfs1)

df1 = pd.DataFrame({'Red': r1, 'Green': g1, 'Blue': b1})
df2 = pd.DataFrame({'Red': r2, 'Green': g2, 'Blue': b2})
```

```python
just_two_row_plot(df1, df2, 'Blue')
```

```python
just_two_row_plot(df1, df2, 'Red')
```

```python
just_two_row_plot(df1, df2, 'Green')
```
